```python
#plt.scatter(x , y, color='color_name', marker='marker style'  ,label ='label name')

import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores= [50,55,60,70,75,80,85,100]

#scatter plot
plt.scatter(hours_studies, exam_scores , color='green' , marker ='o' , label='Student data')

plt.xlabel('Hours studies')
plt.ylabel('Exam Score')
plt.title('Relationship between Study time and exam score')
plt.legend()
plt.grid()
plt.show()
```

```python
#plt.subplot(nrows, ncols , index)

import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) #1 row , 2colum, 1st suplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2) #1row, 2column , 2nd subplot
plt.bar(x,y)
plt.title("Bar chart")

plt.show()

```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

#fig , ax = plt.subplots(nrows , ncols , figsize=(width , height))
fig , ax = plt.subplots(1, 2, figsize=(10 , 5))

x= [1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y , color='pink')
ax[0].set_title('Line plot')

ax[1].bar(x,y, color ='blue')
ax[1].set_title('Bar chart')

fig.suptitle("comparion of bar and line charts")

plt.tight_layout()

plt.show()
```
